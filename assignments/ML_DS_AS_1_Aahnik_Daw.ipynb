````python
cars_df.isnull().sum()
````

````python
cars_df.head()
````

````python
cars_df.dtypes
````

````python
for col in cars_df.columns:
    if cars_df[col].dtype == "object":
        # not possible to compute mean for object data type
        cars_df[col] = cars_df[col].fillna(cars_df[col].mode()[0])
    else:
        cars_df[col] = cars_df[col].fillna(cars_df[col].mean())
````

````python
cars_df.isnull().sum() # prove that all nan are removed
````
