```python
# encoding with one hot encoding

#import one hot encoder
from sklearn.preprocessing import OneHotEncoder

#instantiate the encoder
encoder = OneHotEncoder(drop="first")

#fit it for all cols in cat_cols
encoder.fit(ldf[cat_cols])

#transform the data
encoded_data = encoder.transform(ldf[cat_cols])

ldf_cat_enc = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(cat_cols))
ldf_cat_enc.head()
```

```python
# scale the data using standard scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(ldf[num_cols])
scaled_data = scaler.transform(ldf[num_cols])
nldf = pd.DataFrame(scaled_data,columns=num_cols)
nldf.head()
```

```python
X = pd.concat([nldf, ldf_cat_enc_tfree], axis = 1)
X.shape
```

```python
X.head()
```

```python
y = ldf_cat_enc["loan_status_Fully Paid"]
y
```

```python
from sklearn.model_selection import train_test_split

def split_feature_target(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=125)
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LogisticRegression


def make_logistic_regressor(X,y):
    X_train, X_test, y_train, y_test = split_feature_target(X,y)
    logistic_regressor = LogisticRegression(max_iter=1000)
    logistic_regressor.fit(X_train,y_train)
    print("Score: ",logistic_regressor.score(X_test,y_test))
    return logistic_regressor,X_train,X_test,y_train,y_test

```

```python
X.head()
```

```python
logistic_regressor,X_train,X_test,y_train,y_test = make_logistic_regressor(X,y)
```

```python
logistic_regressor.coef_
```

```python
logistic_regressor.intercept_
```

```python
from sklearn.metrics import roc_curve,roc_auc_score

# generate a no skill prediction (majority class)
ns_probs = [0 for _ in range(len(y_test))]
predicted = logistic_regressor.predict_proba(X_test)
lr_probs = predicted[:,1]
```

```python
ns_auc = roc_auc_score(y_test,ns_probs)
lr_auc = roc_auc_score(y_test,lr_probs)
print("No Skill: ",ns_auc)
print("Logistic Regression", lr_auc)
```

```python
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
y_predicted = logistic_regressor.predict(X_test)
```

```python
from sklearn.metrics import precision_recall_curve,f1_score,auc
lr_precision, lr_recall, _ = precision_recall_curve(y_test, lr_probs)
lr_f1, lr_auc = f1_score(y_test, y_predicted), auc(lr_recall, lr_precision)
# summarize scores
print('Logistic: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
# plot the precision-recall curves
no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
plt.plot(lr_recall, lr_precision, marker='.', label='Logistic')
# axis labels
plt.xlabel('Recall')
plt.ylabel('Precision')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
# Thinking from a bank's perspective, which metric should our primary focus be on..
# - ROC AUC
# - Precision
# - Recall  
# - F1 Score

# here positive prediction => predict that money will be fully paid => bank will proceed to give loan
# and negative prediction => borrower will be unable to pay back full money, and bank need to incurr loss


# if there are lot of false positives, then lot of loans will need to be charged off, and the bank will incurr loss
# if there are lot of false negatives, then the bank would loose lot of potential people who could pay back loan with interest

# ROC = Receiver operator characteristic graphs ==> summarizes the stats from all the confusion matrices that each threshhold produced
# AUC = area under ROC curve. higher the auc, better the model. auc helps us compare models created by different algorithms
# Precision measures how good our model is when the prediction is positive. Recall measures how good our model is at correctly predicting positive classes.

# we need to increase both precision and recall
# precision is directly proportional to True Positive
# recall is directly proportional to True Negative
# if true positives increase then false negatives will decrease
# if true negatives increase then false positives will decrease
# thus we want to increase TP and TN ==> precision and recall
# F1 score is a metric that combines both precision and recall with the same weightage
# so f1 score is the best metric
```

```python
# How does the gap in precision and recall affect the bank?
# if precision is high and recall is low => TP is high and FP is high. the bank will get lot of good loans as well as bad loans

# if precision is low and recall is high
# bank will give less loans. thus it will miss out on many potential good customers,and also avoid many defaulters

# the gap between precision and recall will cause loss for bank
# a model with high precision and high recall is good for the bank
```
